## **Kesimpulan Model Logistic Regression**

Model Logistic Regression adalah model statistik yang digunakan untuk masalah klasifikasi. Cara kerja model Logistic Regression adalah dengan menggunakan fungsi logistik (sigmoid) untuk mengubah hasil dari kombinasi linear fitur-fitur menjadi nilai antara 0 dan 1, yang dapat diinterpretasikan sebagai probabilitas. Jika probabilitas lebih besar dari 0.5, model akan mengklasifikasikan instance ke dalam kelas 1; jika probabilitas lebih kecil atau sama dengan 0.5, instance akan diklasifikasikan ke dalam kelas 0. Hasil dari evaluasi kinerja model Logistic Regression ditampilkan dalam metrik seperti precision, recall, F1-score, dan akurasi.

1. Precision:

> - Precision adalah rasio positif benar (true positive) dibagi dengan total prediksi positif (true positive + false positive).
> - Untuk kelas 0 (label 0), precision adalah 0.82, yang berarti dari semua prediksi yang diklasifikasikan sebagai kelas 0, 82% benar-benar adalah kelas 0.
> - Untuk kelas 1 (label 1), precision adalah 0.64, yang berarti dari semua prediksi yang diklasifikasikan sebagai kelas 1, 64% benar-benar adalah kelas 1.

2. Recall (Sensitivity):

> - Recall adalah rasio positif benar (true positive) dibagi dengan total jumlah data aktual yang sebenarnya adalah kelas positif (true positive + false negative).
> - Untuk kelas 0 (label 0), recall adalah 0.97, yang berarti bahwa dari semua data yang sebenarnya adalah kelas 0, model berhasil mengidentifikasi 97% dari mereka.
> - Untuk kelas 1 (label 1), recall adalah 0.21, yang berarti bahwa dari semua data yang sebenarnya adalah kelas 1, model hanya berhasil mengidentifikasi 21% dari mereka.

3. F1-Score:

> - F1-score adalah rata-rata harmonik dari precision dan recall. Ini memberikan nilai tunggal yang mencakup baik precision maupun recall.
> - F1-score untuk kelas 0 adalah 0.89, dan untuk kelas 1 adalah 0.31.

4. Support:

> Support adalah jumlah data aktual yang termasuk dalam setiap kelas.

5.Accuracy:

> - Accuracy adalah rasio prediksi yang benar (true predictions) dibagi dengan total jumlah prediksi.
> - Accuracy adalah 0.81, yang berarti model Anda memiliki tingkat keakuratan sekitar 81% pada dataset ini.

6. Confusion Matrix:

> - Confusion matrix menunjukkan perbandingan antara prediksi model dengan kelas sebenarnya.
> - Diagonal utama dari matriks ini adalah true positive (1921) dan true negative (108), yang berarti model benar-benar memprediksi dengan benar kelas 0 sebanyak 1921 kali dan kelas 1 sebanyak 108 kali.
> - Diagonal yang tidak utama adalah false positive (62) dan false negative (409), yang berarti model salah memprediksi kelas 0 sebagai kelas 1 sebanyak 62 kali dan kelas 1 sebagai kelas 0 sebanyak 409 kali.

Kesimpulannya, model ini memiliki kinerja yang baik dalam mengklasifikasikan kelas 0 tetapi memiliki kesulitan dalam mengklasifikasikan kelas 1. Diperlukan upaya lebih lanjut, seperti penyesuaian hyperparameter atau penggunaan metode lain, untuk meningkatkan kinerja model terutama pada kelas 1 jika itu adalah tujuan utama dalam konteks aplikasi yang dihadapi.

## Tarik Kesimpulan Model Mana yang terbaik beserta alasannya

```python
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

# Bagi data menjadi data pelatihan dan data pengujian
X_train,X_test,y_train,y_test = train_test_split(X_transform,y,test_size = 0.25,random_state = 123)

# Inisialisasi model-model yang akan diuji
models = [
    ("Random Forest", RandomForestClassifier()),
    ("Gradient Boosting", GradientBoostingClassifier()),
    ("Logistic Regression", LogisticRegression()),
    ("SVM", SVC()),
    ("K-Nearest Neighbors", KNeighborsClassifier()),
    ("Decision Tree", DecisionTreeClassifier())
]

# Evaluasi dan pilih model terbaik
best_model = None
best_accuracy = 0

for name, model in models:
    # Lakukan validasi silang dengan model
    scores = cross_val_score(model, X_train, y_train, cv=10, scoring='accuracy')
    avg_accuracy = scores.mean()

    # Cetak hasil validasi silang
    print(f"{name}: {avg_accuracy:.4f}")

    # Periksa apakah model ini lebih baik dari yang sebelumnya
    if avg_accuracy > best_accuracy:
        best_model = model
        best_accuracy = avg_accuracy

# Latih model terbaik dengan data pelatihan
best_model.fit(X_train, y_train)

# Lakukan prediksi dengan model terbaik pada data pengujian
y_pred = best_model.predict(X_test)

# Evaluasi model terbaik pada data pengujian
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy on test data: {accuracy:.4f}")

# Model terbaik adalah best_model
print("Best Model:", best_model)
```

Jadi, berdasarkan hasil pengujian yang telah dilakukan, diperoleh bahwa nilai akurasi model yang paling tinggi adalah (1) Gradient Boosting dengan akurasi 86,4%, (2) Random Forest dengan akurasi 86,1%, dan (3) Support Vector Machine (SVM) dengan akurasi 85,8%.
